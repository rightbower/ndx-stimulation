```python
import string

def caesar(mystr, ind):
    alpha = string.ascii_lowercase
    newalpha = alpha[ind:]+alpha[0:ind]
    newdict = {alpha[i]: newalpha[i] for i in range(len(alpha))}  # can also use str.maketrans(string1, string2)
                                                                  # this automatically ignores non-alpha chars
    newstr = ''   
    for i in range(len(mystr)):
        if alpha.find(mystr[i]) != -1: # or if mystr[i] in alpha
            newstr += newdict[mystr[i]]
        else:
            newstr += mystr[i]
    return newstr
    
    
caesar('zab!', 4)
```

```python
# p-code: iterate thru list, pairwise element comparisons and swap if out of order
#         continue iterating thru list until have passed through whole list with no swaps

def bubble_sort(lst):
    n = len(lst)
    swapped = True
    while swapped:
        swapped = False
        for i in range(n-1):
            if lst[i]>lst[i+1]:
                small = lst[i+1]
                lst[i+1]=lst[i]
                lst[i]=small
                swapped = True
    return

# what about duplicates? already handled - no need to swap if same
                
lst = [9, 3, 2, 6, 7]
print(lst)
bubble_sort(lst)
print(lst)
```

```python

def merge_sort(lst):
    
    n = len(lst)
    if n<=1:
        return lst
    
    mid = n//2
    left = merge_sort(lst[:mid])
    right = merge_sort(lst[mid:])
    
    return merge(left, right)


def merge(left, right):
    
    result = []
    while (left != []) and (right != []):
        if left[0] <= right[0]:
            result.append(left[0])
            left = left[1:]
        else:
            result.append(right[0])
            right = right[1:]
            
    while left!=[]:
        result.append(left[0])
        left = left[1:]
        
    while right!=[]:
            result.append(right[0])
            right = right[1:]
            
    return result
        

```

```python
# this is the worst way to pick a pivot - see wikipedia page for better pivots (0, mid, end vals and take middle)

def quicksort(lst):
    
    if lst == []:
        return []
    else:
        pivot = lst[0]
        result = [pivot]
    
        if len(lst)==1:
            return result
        else:
            left = []
            right = []
            for i in lst[1:]:
                if i<pivot:
                    left.append(i)
                else:
                    right.append(i)
        left = quicksort(left)
        right = quicksort(right)
    
    return left + result + right
    
lst = [1, 3, 2, 2]
yay = quicksort(lst)
print(yay)
```

```python
# Binary heap: complete BT (above) with heap property - each child of parent has key is >= (min) or <= (max) parent
#              max/min of heap is root node, min/max is min/max value from deepest level - depth-traversal but only
#              store keys from lowest level to find min/max
#            : tree or array structure (since complete) - array is simple, use indexing (node n - children 2n, 2n+1)
#
# heapsort: 1) swap root with final value in array 2) sift_down root (swap root with child nodes if fails heap prop)
#           3) recursively apply procedure on array[0:n-1] since last element is now max
#         : can use heapsort to find k smallest/largest values in a list (heapify list to bin heap array)
#
# REMEMBER:
# Heap structure (FULL tree) vs heap order (each level all > (min) or < (max) than prev)
# can use percUp function on heap structure to restore order. this is why replace root with last list value and percUp
#
# def heapify(lst, count):
    # takes a list of values and puts it in binary heap form [root, child_l, child_r, childl_l, childl_r, ...]
    # try implementing sift_down vs sift_up method? (see binary heap wikipedia)
    
def percUp(lst, i):
    # heapify list of keys
    while i//2 > 0:            # if it's zero, that's root
        if lst[i] < lst[i//2]: # list parent greater than child (this is min heap)
            tmp = lst[i//2]
            lst[i//2] = lst[i]
            lst[i] = tmp
        i = i//2
        
def insert(lst, k):
    # insert new key in list and percUp till in correct place
    lst.append(k)
    i = len(lst)
    percUp(lst, i)
    

def percDown(lst, i):
    # shift node down lst until found right level; i is current position
    # swap with smallest child until node key is < both children (> for max)
    
    while 2*i < len(lst):
        # find index minchild of i
        if (2*i)+1 > len(lst):
            minchild = 2*i
        elif lst[2*i] < lst[(2*i)+1]:
            minchild = 2*i
        else:
            minchild = 2*i+1
            
        if lst[i] > lst[minchild]:
            tmp = lst[minchild]
            lst[minchild] = lst[i]
            lst[i] = tmp
        
        
        i = minchild
            
            
def delmin(lst):
    # delete min i.e. root node, return min val
    retval = lst[0]
    lst[0] = lst[-1]
    lst.pop()
    percDown(lst, 0)
    return retval


def buildheap(lst):
    #
    i = len(lst) // 2
    heaplist = lst[:]
    
    while i>0:
        percDown(heaplist, i)
        i = i-1
        
    return heaplist
        
        
        
lst = [10, 1, 2, 3, 7, 8]
percDown(lst, 0)
print(lst)
```

```python
# Binary search: on sorted list, divide and conquer (check mid value, if < bin sort on right array, > left)
#
# Binary search tree: INORDER TRAVERSAL OF BINARY TREE GIVES SORTED LIST IF PUSH KEYS TO STACK
# 

class TreeNode():
    
    def __init__(self, key, value=None, left = None, right = None):
        self.key = key
        self.value = value
        self.left = left
        self.right = right
        

class BinarySearchTree():
    
    def __init__(self):
        self.root = None
        
    def put(self, key, value):
        if self.root:
            self._put(key, value, self.root)
        else:
            self.root = TreeNode(key, value)
                        
    def _put(self, key, value, node):
        if key < node.key:
            if node.left:
                self._put(key, value, node.left)
            else:
                node.left = TreeNode(key, value)
        else:
            if node.right:
                self._put(key, value, node.right)
            else:
                node.right = TreeNode(key, value)
                
    def __setitem__(self, key, value):
        self.put(key, value)
        
        
    def _list(self, node, indent=0):
        assert node, "invalid node"
        if node:
            print(" " * indent, node.key,"=", node.value)
            if node.left:
                print(" "*indent, "<", end="")
                self._list(node.left, indent+1)
            if node.right:
                print(" "*indent, ">", end="")
                self._list(node.right, indent+1)
                
                
    def list(self, start=""):
        print("\n\n----", start)
        self._list(self.root)
        
    
        
            
#lst = [6, 3, 8, 5, 9, 2, 7, 1]
lst = [4, 2, 5, 3, 1]
yay = BinarySearchTree()
for x in lst:
    yay.put(x, None)
yay.list()
```
